```
lst = range(0,100,10)
for element in lst:
  print(f'element - {element}')
```

```
lst = range(0,100,10)
for i,element in enumerate(lst):
  print(f'element index {i} is {element}')
```

```
covid_19_data_Confirmed = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
covid_19_data_Recovered = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"
covid_19_data_Deaths = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
```

```
## read csv online:
df_covid_confirmed = pd.read_csv(covid_19_data_Confirmed)
## read csv online:
df_covid_Recovered = pd.read_csv(covid_19_data_Recovered)
## read csv online:
df_covid_Deaths = pd.read_csv(covid_19_data_Deaths)
```

```
df_covid = df_covid_confirmed
df_covid.head()
```

```
## Get all countries
Countries = df_covid['Country/Region'].unique()
Countries
```

```
## Get Spain, Germany and France data
df_covid_c = df_covid[df_covid['Country/Region'].isin(['Spain','Germany','Poland','Belgium'])].copy()
df_covid_c.head(20)
```

```
## keep the ones which Province/Sate == NaN
df_covid_c = df_covid_c[df_covid_c['Province/State'].isnull()]
df_covid_c.T
```

```
df_covid = df_covid_confirmed
df_covid.head()
```

```
## Get all countries
Countries = df_covid['Country/Region'].unique()
Countries
```

```
## let us plot some data now
```

```
df_covid_c = df_covid_c.T.iloc[4:]
df_covid_c.columns = ['Belgium','Germany','Poland','Spain']
df_covid_c.tail(10)
```

```
## conver from cumulative to daily
df_covid_c = df_covid_c.diff().fillna(0)
df_covid_c = df_covid_c.rolling(window = 14).mean().fillna(0)
```

```
## let us check the column types
df_covid_c.dtypes
```

```
## let us check the column types
df_covid_c.dtypes
```

```
## let us convert Date to pandas datetime object
df_covid_c.index = pd.to_datetime(df_covid_c.index)
```

```
df_covid_c.head(10)
```

```
## import library
import matplotlib.pyplot as plt
```

##Line plot

```
f, ax = plt.subplots(1,1,figsize=(10,7))
ax.plot(df_covid_c.index,df_covid_c.Spain,label = 'Spain')
ax.plot(df_covid_c.index,df_covid_c.Poland,label = 'Poland')
ax.plot(df_covid_c.index,df_covid_c.Belgium,label = 'Belgium')
ax.plot(df_covid_c.index,df_covid_c.Germany,label = 'Germany')
ax.legend()
```

```
df_covid_c.plot()
```

##scatter plot

```
df_covid_c.reset_index(inplace=True)
df_covid_c.columns = ['Date','Belgium','Germany','Poland','Spain']
df_covid_c.head(10)
```

```
df_covid_c.dtypes
```

```
colors = df_covid_c.index
f,ax = plt.subplots(1,1,figsize=(10,7))
sc = ax.scatter(df_covid_c['Germany'],
                df_covid_c['Spain'],
                c=colors)
ax.set_xlim([0,max(df_covid_c['Spain'].max(),df_covid_c['Germany'].max())])
ax.set_ylim([0,max(df_covid_c['Spain'].max(),df_covid_c['Germany'].max())])
ax.set_ylabel('Spain')
ax.set_xlabel('Germany')
plt.colorbar(sc)
```

```
f,ax=plt.subplots(2,2,figsize=(10,7),sharey=True)
ax[0][0].plot(df_covid_c['Date'],df_covid_c['Germany'])
ax[0][0].set_title('Germany')

ax[0][1].plot(df_covid_c['Date'],df_covid_c['Belgium'])
ax[0][1].set_title('Belgium')

ax[1][0].plot(df_covid_c['Date'],df_covid_c['Poland'])
ax[1][0].set_title('Poland')

ax[1][1].plot(df_covid_c['Date'],df_covid_c['Spain'])
ax[1][1].set_title('Spain')
f.autofmt_xdate(rotation=45)
```

```
## bar chart
f,ax = plt.subplots(1,1,figsize=(10,8))
df_covid_14Day = df_covid_c.iloc[-14:].copy()

df_covid_14Day['Date'] = df_covid_14Day['Date'].dt.strftime('%D')
df_covid_14Day.plot(kind = 'bar',
                                  stacked = True,
                                  x = 'Date',
                                  y = ['Belgium','Germany','Poland','Spain'],
                                  ax=ax)
ax.set_ylabel('Confirmed cases', fontsize = 14);
ax.set_xlabel('Dates',fontsize = 14);
```
